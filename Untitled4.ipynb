```python
import numpy as np

def generar_saldo():
    return int(np.random.choice(np.linspace(100000, 1000000, 100))) #genero un saldo inicial, el linspace genera 100 valores entre 100mil y 1 millon y el random elige uno de esos valores para tomarlo
                                                                    #como saldo inicial

def menu():
    print("\n=== CAJERO AUTOMÁTICO ===")        #opciones que se elegiran en la funcion 'cajero_automatico', posteriormente defino cada una en funciones
    print("1. Consultar saldo")
    print("2. Depositar")
    print("3. Retirar")
    print("4. Transferir dinero")
    print("5. Salir")

def consultar_saldo(saldo):                               #opc 1
    print(f"Saldo actual: ${saldo}")

def depositar(saldo):                                    #opc 2
    monto = int(input("Ingrese el monto a depositar: "))
    if monto > 0:
        saldo += monto
        print(f"Depósito exitoso. Nuevo saldo: ${saldo}")
    else:
        print("Monto inválido.")
    return saldo

def retirar(saldo):                                   #opc 3
    monto = int(input("Ingrese el monto a retirar: "))
    if 0 < monto <= saldo:
        saldo -= monto
        print(f"Retiro exitoso. Nuevo saldo: ${saldo}")
    else:
        print("Monto inválido o saldo insuficiente.")
    return saldo

def transferir(saldo):
    cuenta = input("Ingrese el número de cuenta de destino: ")
    monto = int(input("Ingrese el monto a transferir: "))
    if 0 < monto <= saldo:
        saldo -= monto
        print(f"Transferencia de ${monto} a la cuenta {cuenta} realizada.")
        print(f"Nuevo saldo: ${saldo}")
    else:
        print("Monto inválido o saldo insuficiente.")
    return saldo

def ingresar_contraseña():
    contraseña = "1234"
    intentos = 3                      #pedir una contraseña por entrada y compararla con una contraseña ya establecida '1234'
    while intentos > 0:
        ingreso = input("Ingrese su contraseña: ")
        if ingreso == contraseña:
            print("Acceso concedido.")
            return True
        else:
            intentos -= 1
            print(f"Contraseña incorrecta. Intentos restantes: {intentos}")
    print("Acceso denegado.")
    return False

def cajero_automatico():
    saldo = generar_saldo()
    if not ingresar_contraseña(): # esta negando el booleano que da en ingresar contraseña. si el resultado de la negacion es true entra al if y no deja ingresar y el return detiene la ejecucion de func.
        return

    while True:
        menu()
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            consultar_saldo(saldo)
        elif opcion == "2":
            saldo = depositar(saldo)
        elif opcion == "3":
            saldo = retirar(saldo)
        elif opcion == "4":
            saldo = transferir(saldo)
        elif opcion == "5":
            print("Gracias por usar el cajero automático. ¡Adiós!")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

cajero_automatico()
```
