#Ejercicio 1: Cilindro inscrito en un cono

$V = \pi r^2 h$

$\frac{r}{H - h} = \frac{R}{H} \Rightarrow r = \frac{R}{H}(H - h)$

$V(h) = \pi \left( \frac{R}{H}(H - h) \right)^2 h = \pi \frac{R^2}{H^2}(H - h)^2 h$

$V'(h) = \pi \frac{R^2}{H^2} \left[ -2(H - h)h + (H - h)^2 \right]$

$= \pi \frac{R^2}{H^2}(H - h)[(H - h) - 2h] = \pi \frac{R^2}{H^2}(H - h)(H - 3h)$

Puntos críticos: $h = H$ (borde), $h = H/3$

$h = \frac{H}{3}, \quad r = \frac{R}{H}(H - H/3) = \frac{2R}{3}$

El cilindro de volumen máximo tiene altura=$H/3$ y radio=$2R/3$

# Ejercicio 2: Energía mínima al nadar río arriba

- Velocidad del pez respecto al agua: $v$ \\
- Velocidad de la corriente: $v_c$ \\
- Distancia: $d$ \\
- Energía: proporcional a $t v^3$

$t = \frac{d}{v - v_c}, \quad E = k t v^3 = k \frac{d v^3}{v - v_c}$

$E'(v) = k d \cdot \frac{(3v^2)(v - v_c) - v^3}{(v - v_c)^2}
= k d \cdot \frac{2v^3 - 3v^2 v_c}{(v - v_c)^2}$

$2v^3 - 3v^2 v_c = 0 \Rightarrow v^2(2v - 3v_c) = 0 \Rightarrow v = \frac{3}{2} v_c$

La velocidad óptima es $v = \frac{3}{2} v_c$

###Ejercicio

En la fabricación y venta de x unidades de cierto bien de consumo, las funciones de precio $P$ y de costo $C$ (en dólares) están dadas por:

$$P(x) = 5.00 - 0.002x$$

$$C(x)=3.00 + 1.10x$$

Encuentre las expresiones para el ingreso, el costo y la utilidad marginales. Determine el nivel de producción que producirá la máxima utilidad total.Explique con claridad cada proceso

Solucion:

## Funciones dadas

$$
P(x) = 5 - 0.002x \quad \Rightarrow \quad P'(x) = -0.002
$$

$$
C(x) = 3 + 1.10x \quad \Rightarrow \quad C'(x) = 1.10
$$

$$
I(x) = P(x) \cdot x = (5 - 0.002x)x = 5x - 0.002x^2
$$

$$
I'(x) = 5 - 0.004x
$$

$$
U(x) = I(x) - C(x) = (5x - 0.002x^2) - (3 + 1.10x)
$$

$$
U(x) = 5x - 0.002x^2 - 3 - 1.10x = -0.002x^2 + 3.9x - 3
$$

$$
U'(x) = -0.004x + 3.9
$$

---

## Hallar el máximo

$$
U'(x) = 0 \Rightarrow -0.004x + 3.9 = 0
$$

$$
-0.004x = -3.9 \Rightarrow x = \frac{-3.9}{-0.004} = 975
$$

---

## Verificación de máximo

$$
U'(970) = 3.9 - 0.004 \cdot 970 = 0.02 \quad \text{(positivo)}
$$

$$
U'(980) = 3.9 - 0.004 \cdot 980 = -0.02 \quad \text{(negativo)}
$$

$$
\Rightarrow \text{En } x = 975 \text{ la derivada cambia de positiva a negativa, por lo tanto es un máximo.}
$$

---

### Conclusión

$$
\boxed{\text{En 975 unidades, la utilidad alcanza el máximo.}}
$$

```python
#II.
def f(x):
    return x**2 + x**4

x = np.linspace(-2, 2, 400)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$f(x) = x^2 + x^4$", color="blue")
plt.axhline(0, color='black', linewidth=0.5)  # Eje x
plt.axvline(0, color='black', linewidth=0.5)  # Eje y
plt.scatter([0, 1, -1], [f(0), f(1), f(-1)], color='red', zorder=5)
plt.text(1, f(1)+0.2, "(1, 2)", ha='center')
plt.text(-1, f(-1)+0.2, "(-1, 2)", ha='center')
plt.text(0, f(0)-0.5, "(0, 0)", ha='center')
plt.title("Bosquejo de una función que cumple las condiciones")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.ylim(-1, 5)

plt.show()
```

```python
#III.
from scipy.interpolate import CubicSpline

x_points = [2, 4, 6]
y_points = [-3, -1, 1]
cs = CubicSpline(x_points, y_points, bc_type=((1, 0.0), (1, 3.0)))
x = np.linspace(1, 8, 400)
y = cs(x)
y_prime = cs.derivative(1)(x)
y_double_prime = cs.derivative(2)(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x)", color="blue")
plt.plot(x, y_prime, '--', label="f'(x)", color="green")
plt.plot(x, y_double_prime, '--', label="f''(x)", color="red")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.scatter([2, 6], [-3, 1], color='purple')
plt.text(2, -3.3, "(2, -3)", ha='center')
plt.text(6, 1.2, "(6, 1)", ha='center')
plt.title("Bosquejo de una función con las condiciones dadas")
plt.xlabel("x")
plt.ylabel("f(x), f'(x), f''(x)")
plt.grid(True)
plt.legend()
plt.ylim(-5, 5)

plt.show()
```
