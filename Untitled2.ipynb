```python
#suma de los dos numeros anteriores--- eje: 0,1,1,2,3,5,8,13,21-- 0+1=1,1+1=2,1+2=3
#Escribe un programa en Python que genere los primeros n números de la serie de Fibonacci
#y visualízalos en un gráfico de líneas utilizando Matplotlib. El usuario debe poder elegir el número n.
import matplotlib.pyplot as plt
def fibonacci(n):
    serie = [0, 1]
    for i in range(2, n):                      #i=2   i=[2-1] + i=[2-2]
        serie.append(serie[i-1] + serie[i-2]) #serie(2)=  1   +   0       ---> es decir que cuando x vale 2, y vale 1
    return serie[:n] #devuelve la cantidad exacta, por ejemplo si pide 1, pues en teoria devolveria la lista inicia [0,1], pero especificando devolverá el 0.
cant=int(input("digite la cantidad de numeros de fibonnaci: "))
fibonacci(cant)
print(fibonacci(cant))
plt.plot(fibonacci(cant), marker= 'o') #graafica linea (serie, marcador)
plt.title("Fibonacci") #titulo grafica
plt.xlabel("numero") #titulo eje y
plt.ylabel("valor") #titulo eje x
plt.grid(True)  #cuadricula
plt.show() #muestra la gradfica
```
