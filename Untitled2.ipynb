```python
#1) suma de los dos numeros anteriores--- eje: 0,1,1,2,3,5,8,13,21-- 0+1=1,1+1=2,1+2=3
#Escribe un programa en Python que genere los primeros n números de la serie de Fibonacci
#y visualízalos en un gráfico de líneas utilizando Matplotlib. El usuario debe poder elegir el número n.
import matplotlib.pyplot as plt
def fibonacci(n):
    serie = [0, 1]
    for i in range(2, n):                      #i=2   i=[2-1] + i=[2-2]
        serie.append(serie[i-1] + serie[i-2]) #serie(2)=  1   +   0       ---> es decir que cuando x vale 2, y vale 1
    return serie[:n] #devuelve la cantidad exacta, por ejemplo si pide 1, pues en teoria devolveria la lista inicia [0,1], pero especificando devolverá el 0.
cant=int(input("digite la cantidad de numeros de fibonnaci: "))
fibonacci(cant)
print(fibonacci(cant))
plt.plot(fibonacci(cant), marker= 'o') #graafica linea (serie, marcador)
plt.title("Fibonacci") #titulo grafica
plt.xlabel("numero") #titulo eje y
plt.ylabel("valor") #titulo eje x
plt.grid(True)  #cuadricula
plt.show() #muestra la gradfica
```

```python
#2) Dado un conjunto de datos con las temperaturas diarias de una ciudad durante un año, escribe un programa que calcule la media, la mediana
# y la desviación estándar de las temperaturas. Luego, crea un gráfico de líneas que muestre la temperatura media mensual.
#temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]
import matplotlib.pyplot as plt
import random
temperaturas =  [random.randint(20, 35) for i in range(365)]
print("temp: " , temperaturas)
def calcularMedia(temperaturas):
    return sum(temperaturas)/len(temperaturas)
def calcularMediana(temperaturas):
    temperaturas.sort()
    if len(temperaturas) % 2 == 0:
        return (temperaturas[len(temperaturas)//2-1] + temperaturas[len(temperaturas)//2])/2
    else:
        return temperaturas[len(temperaturas)//2]
def calcularDesviacionEstandar(temperaturas):
    media = calcularMedia(temperaturas)
    suma = sum([(x - media)**2 for x in temperaturas])
    return (suma/(len(temperaturas)-1))**0.5
media = calcularMedia(temperaturas)
mediana = calcularMediana(temperaturas)
desviacionEstandar = calcularDesviacionEstandar(temperaturas)
print("temp ordenadas" ,temperaturas)
print("Media:", media)
print("Mediana:", mediana)
print("Desviación estándar:", desviacionEstandar)
temperaturasMensuales = [sum(temperaturas[i:i+30])/30 for i in range(0,360, 30)]
print("temp mensuales:" , temperaturasMensuales)
plt.plot(temperaturasMensuales, marker=".")
plt.title("Temperaturas mensuales")
plt.xlabel("Mes")
plt.ylabel("Temperatura media")
plt.grid(True)
plt.show()

```

```python
#4) Escribe un programa que grafique las funciones seno, coseno y tangente en el mismo plano para valores entre 0 y 2pi.
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 1000)

y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

plt.figure(figsize=(8, 6)) # tamaño imagen (ancho, alto)
plt.plot(x, y_sin, label='Seno', color='b')
plt.plot(x, y_cos, label='Coseno', color='r')
plt.plot(x, y_tan, label='Tangente', color='g', linestyle='dashed')
plt.ylim(-10,10)
plt.xlabel('valores x (radianes)')
plt.ylabel('Valor de la función')
plt.title('Funciones Trigonométricas: Seno, Coseno y Tangente')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend() #incica cual funcion es cual (tan ---, sen...) etiqueta
plt.grid()
plt.show()
```

```python
#6) Pendiente de una recta: Escribe un programa que tome dos puntos en el plano (x1, y1) y (x2, y2), calcule la pendiente de la recta que los une,
# y grafique dicha recta.

import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

def calcular_pendiente(x1, y1, x2, y2):
    if x2 - x1 == 0:
        return None  # Pendiente indefinida (recta vertical)
    return (y2 - y1) / (x2 - x1)

def graficar_recta(x1, y1, x2, y2):
    # Crear un rango de valores x para graficar la recta
    x_values = [x1, x2]
    y_values = [y1, y2]

    plt.plot(x_values, y_values, marker='o')  # Graficar los puntos
    plt.title('Recta que une los puntos')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid()
    plt.axhline(0, color='black',linewidth=0.5, ls='--')
    plt.axvline(0, color='black',linewidth=0.5, ls='--')
    plt.xlim(min(x1, x2) - 1, max(x1, x2) + 1)
    plt.ylim(min(y1, y2) - 1, max(y1, y2) + 1)
    plt.show()

def main():
    # Solicitar puntos al usuario
    x1 = float(input("Ingrese x1: "))
    y1 = float(input("Ingrese y1: "))
    x2 = float(input("Ingrese x2: "))
    y2 = float(input("Ingrese y2: "))

    # Calcular la pendiente
    pendiente = calcular_pendiente(x1, y1, x2, y2)

    if pendiente is None:
        print("La pendiente es indefinida (la recta es vertical).")
    else:
        print(f"La pendiente de la recta que une los puntos es: {pendiente}")

    # Graficar la recta
    graficar_recta(x1, y1, x2, y2)

if __name__ == "__main__":
    main()


```

```python
#7) Movimiento parabólico: Escribe un programa que simule el movimiento de un proyectil bajo la influencia de la gravedad (sin resistencia del aire).

#Grafica la trayectoria parabólica para diferentes ángulos de lanzamiento.

import numpy as np
import matplotlib.pyplot as plt

def trayectoria_proyectil(v0, theta, g=9.81):
    theta_rad = np.radians(theta)

    t_flight = (2 * v0 * np.sin(theta_rad)) / g

    t = np.linspace(0, t_flight, num=500)

    x = v0 * np.cos(theta_rad) * t
    y = v0 * np.sin(theta_rad) * t - 0.5 * g * t**2

    return x, y

def graficar_trayectorias(v0, angulos):
    plt.figure(figsize=(10, 6))

    for theta in angulos:
        x, y = trayectoria_proyectil(v0, theta)
        plt.plot(x, y, label=f'Ángulo: {theta}°')

    plt.title('Trayectoria de un proyectil')
    plt.xlabel('Distancia (m)')
    plt.ylabel('Altura (m)')
    plt.grid()
    plt.axhline(0, color='black', linewidth=0.5, ls='--')
    plt.axvline(0, color='black', linewidth=0.5, ls='--')
    plt.legend()
    plt.xlim(0, 2 * (v0**2) * np.sin(np.radians(max(angulos))) / 9.81)
    plt.ylim(0, (v0**2) * (np.sin(np.radians(max(angulos)))**2) / (2 * 9.81) + 1)
    plt.show()

def main():
    v0 = float(input("Ingrese la velocidad inicial del proyectil (m/s): "))

    angulos = [15, 30, 45, 60, 75]

    graficar_trayectorias(v0, angulos)

if __name__ == "__main__":
    main()
```
